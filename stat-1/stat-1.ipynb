```python
from math import sqrt

n = 250
n1 = n
n2 = n
x1 = 71
x2 = 58

def z_criterion(x1, n1, x2, n2):
    print(
    '''
    Решение задачи о значимости различий
    вычислением Z-критерия
    '''
    )
    
    p1 = x1 / n1
    p2 = x2 / n2

    p12 = (x1 + x2) / (n1 + n2)

    z = abs(p1 - p2) / sqrt(p12 * (1 - p12) * (1 / n1 + 1 / n2))
    
    z_list = {0.05: 1.96}
    a = 0.05

    z_max = z_list[a] # z-критерий для довертельного интервала 95%
    print("z = %.3f, z_max = %.3f" % (z, z_max))

    if z < z_max:
        print("""
Критерий z < z_max для уровня значимости 0.05 => Разность
результатов, полученных на двух выборках, лишь
менее чем с 5% уверенностью может свидетельствовать о влиянии
такого фактора, как шаблон, на отклик пользователя,
поэтому мы остаёмся верны нулевой гипотезе: шаблон не влияет
на совершение адресатом письма целевого действия.
        """)
    else:
        print("""
Критерий z > z_max для уровня значимости 0.05 => Разность
результатов, полученных на двух выборках, не позволяет исключить
влияние шаблона на совершение пользователем целевого действия,
поэтому нам стоит принять во внимание такой фактор влияния на
отклик адресата рассылки, как шаблон письма
        """)
        
def chi2(x1, n1, x2, n2):
    print( 
    '''
    Решение задачи о значимости различий
    методом хи-квадрат (Критерий согласия Пирсона)
    '''
    )
    
    '''
    Вычисляем объединённые значения, что допустимо
    в условиях верности нулевой гипотезы: фактор
    шаблона письма не оказывает влияния на отклик,
    т.е. на обеих выборках проведены одинаковые
    эксперименты
    '''
    x_sum = x1 + x2
    n_sum = n1 + n2
    f_e = x_sum / n_sum
    
    # Наблюдаемые частоты
    f1_o = x1 / n1
    f2_o = x2 / n2
    
    # хи-квадрат с поправкой Йетса
    def chi2sum_discrete(value_list):
        acc = 0
        for f_o, f_e in value_list:
            diff = (abs(f_o - f_e) - 0.5)
            acc += diff * diff / f_e
        return acc
    
    # подсчитаем хи-критерий для двух интервалов
    chi2 = chi2sum_discrete([[f1_o, f_e], [f2_o, f_e]])
    print("chi2: %f" % chi2)
    
    
    # число степеней свободы
    R = 2 # число вариаций исследования
    C = 2 # число параметров исследования
    df = (R - 1) * (C - 1)
    
    from scipy import stats
    
    # вычислим хи-критерий для уровня значимости 0.05
    chi2_max = stats.chi2.ppf(0.95, df = df)
    print("chi2_max: %f" % chi2_max)
    
    if chi2 > chi2_max:
        print("Вышли за порог в 5%")
    else:
        print("Ошибка менее 5%")
    
        
z_criterion(x1, n1, x2, n2)
print("-" * 62)
chi2(x1, n1, x2, n2)
```
